We now have a master data frame listing all 500 ML customers' activity across calls, texts, and data usage. To calculate monthly revenue, we will have to refer to customers' plan entitlements to calculate overages, followed by the fee rate in place with their respective plans

```python
# Calculate the monthly revenue for each user
def calculate_revenue(row):
    #calculating customer usage
    minutes = row['monthly_minutes']
    messages = row['#_messages']
    #converting MB to GB and rounding up
    internet_usage = np.ceil(row['mb_used']/ 1024)

    #Plan Entitlements. What customers are limited to:
    entitled_minutes = row['minutes_included']
    entitled_messages = row['messages_included']
    entitled_internet = row['gb_per_month_included']
    
    #Overage Fee Schedule
    overage_minutes = row['usd_per_minute']
    overage_messages = row['usd_per_message']
    overage_internet = row['usd_per_gb']
    
    #Initializing empty sequence
    revenue_minutes = 0
    revenue_messages = 0
    revenue_internet = 0
    
    if minutes > entitled_minutes:
        revenue_minutes = (minutes - entitled_minutes) * overage_minutes
    if messages > entitled_messages:
        revenue_messages = (messages - entitled_messages) * overage_messages
    if internet_usage > entitled_internet:
        revenue_internet = (internet_usage - entitled_internet) * overage_internet
        
    revenue = revenue_minutes + revenue_messages + revenue_internet + row['usd_monthly_pay']
    return revenue

#we create a new column in merged_ml for revenue
merged_ml['revenue'] = merged_ml.apply(calculate_revenue, axis = 1)
display(merged_ml)
```

The variance of Surf revenue appears to be far greater than that for Ultimate customer revenue. Surf revenue tends to either be low or high, whereas Ultimate revenue deviates far less and is more consistent.
