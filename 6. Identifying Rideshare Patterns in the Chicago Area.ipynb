# Hypothesis Testing

We are tesing the following hypothesis:
- "The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."

H0 Null Hypothesis (=) : The average duration of rides from the Loop to O'Hare International Airport on Saturdays is unaffected by 'Good' or 'Bad' weather conditions.

H1 Alternative Hypothesis : The average duration of rides from the Loop to O'Hare International Airport differs on Saturdays depending on 'Good' or 'Bad' weather conditions.

We will perform a two-tailed t-test on Saturday, as a change in either direction can help us reject the null hypothesis. 

```python
# Filter for good and bad days 
good_days = table_7.loc[table_7['weather_conditions'] == 'Good']
bad_days = table_7.loc[table_7['weather_conditions'] == 'Bad']

# Get ride duration for 
gday_duration = good_days['duration_seconds'].to_numpy() 
                            
# Get sample 2
bday_duration = bad_days['duration_seconds'].to_numpy()

# Initialize the alpha value 0.05
alpha = 0.05

# Run the t-test, passing both filtered variables and assuming variance is not equal
result = st.ttest_ind(
    gday_duration,
    bday_duration,
    equal_var = False)

# Print the p-value
print('p-value: ', result.pvalue)

# if-statement displaying the outcome
if (result.pvalue < alpha):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```

We can handily reject the null hypothesis with a p-value that approaches 0 so closely. This indicates that there is a close to 0 percent chance that weather did not affect the ride duration for the Saturdays we have the data for. We can build on this low p-value, by assuming that weather conditions strongly affect ride duration on days other than Saturdays and cities other than Chicago. 

Ride duration is likely longer for many reasons, as bad weather usually means snow, rain, or ice on roads which lowers speeds as drivers operate more cautiously, creating more traffic. These conditions also cause more accidents by those driving less cautiously, lowering speeds and in some cases blocking freeways, creating more traffic.

Bad weather is anecdotally a boon for cab drivers and our findings back this up. Even if weather does not increase demand for taxi rides in general, demand is likely greater during bad weather as unless the majority of companies have extra drivers on the road, the longer average ride duration will mean fewer drivers able to serve riders. Zuber should consider increasing the number of employees working on rainy and icy days to account for the increased demand, and pricing promotions to nudge the increased demand their way.
